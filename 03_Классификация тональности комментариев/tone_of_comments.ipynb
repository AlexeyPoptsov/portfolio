Сразу оговорюсь, в связи с некоторыми обстоятельствами работа над проектом начата очень поздно, поэтому сам проект будет выполнен по принципу достаточного минимализма, что крайне не свойственно мне (судя по предыдущим проектам), но обстоятельсва вынуждают. :)

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer 

from sklearn.utils import shuffle
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
from lightgbm import LGBMClassifier
import time
```

```python
# Некоторые общие константы

state = 12345
cores = -1
```

```python
# Загрузим и оценим исходные данные

df = pd.read_csv('/datasets/toxic_comments.csv')
df.head()
```

```python
df.info()
```

### Разобьем коменты на твиты с фильтрацией символов. 

```python
# Выполним токенизацию с фильтрацией символов: Английские буквы и символ '

token = RegexpTokenizer('[a-zA-Z\']+')  
df['token'] = df['text'].str.lower().apply(token.tokenize)
df.head()
```

### Удалим стоп слова

```python
stop_words = set(stopwords.words('english'))
```

```python
list_token = df['token']
list_row = []
for row in list_token:
    without_stop_words = [word for word in row if word not in stop_words] 
    list_row.append(without_stop_words)
df['token'] = list_row
```

```python
df.head()
```

### Лемматизируем 

```python
lemmatizer = WordNetLemmatizer()
```

```python
list_token = df['token']
list_row = []
error_list = []
for  i, row in enumerate(list_token):
    try: 
        lemmatized = [lemmatizer.lemmatize(word) for word in row]
    except:
        lemmatized = ''
        print('Ошибка лематизации', row)
        error_list.append(i)
    list_row.append(lemmatized)
    
df['token'] = list_row
```

```python
# В отдельное поле обединим лематизированные токены

df['lemmatized'] = df['token'].apply(lambda x: ' '.join(x))
```

```python
df.head()
```

```python
df.info()
```

Предобработка исходных данных закончена

### Соотношение классов 

```python
def plot_pie_class(df, column):
    df2 =df.groupby(column).size().reset_index(name='counts')
    
    fig, ax = plt.subplots()
    ax.pie(df2['counts'], labels = df2[column], autopct='%1.1f%%' );
    ax.set_title('Соотношение классов {} = 0 \ {} = 1'.format(column, column));
```

### Разделим на трейн и тест выборки

```python
X = df['lemmatized']
y = df['toxic']
```

```python
X.shape, y.shape
```

```python
# Раздление на трейн и тест выборки с перемешиванием и сохранением долей классов.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, shuffle = True, stratify = y, random_state = state)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Скорректируем соотношение классов

Пока не однозначное решение, но корректировать соотношение классов всетаки необходимо. В трейновой выборке. Пока не понятно, исказит ли это словарь TF - IDF так как количество слов всетаки изменится, но соотношения долей слов всетаки сохранятся.

```python
# Функция увеличивает Класс=1 repeat раз.
#
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=state)
    
    return features_upsampled, target_upsampled
```

```python
# Для выравнивания классов до уровня 70\30 увеличим меньший класс в 4 раза.

X_train_upsampled, y_train_upsampled = upsample (X_train, y_train, 4)
```

```python
X_train_upsampled.shape, y_train_upsampled.shape
```

```python
plot_pie_class(pd.DataFrame({'lemmatized':X_train_upsampled, 'toxic':y_train_upsampled}), 'toxic')
```

### Векторизуем методом TF- IDF

Так как данный метод объективно более прогрессивный по сравнению с мешком слов, остановимся только на нем.

```python
# Векторизуем исходные данные. Обучимся на тестовой upsampled выборке. И трансформируем тестовую выборку
vectorizer  = TfidfVectorizer()

X_tf_idf_train = vectorizer.fit_transform(X_train_upsampled)
X_tf_idf_test = vectorizer.transform(X_test)
```

```python
X_tf_idf_train.shape, X_tf_idf_test.shape
```

## Обучение

Ислледуем две принципиально разные модели. Логистическую регрессию и Бустинг над деревьями.

### Модель LogisticRegression

#### Подготовим и обучим модель

```python
model1 = LogisticRegression(penalty = 'l2', random_state = state, n_jobs = cores)
```

```python
model1.get_params()
```

```python
%%time
model1.fit(X_tf_idf_train, y_train_upsampled)
```

#### Оценим метрику F1

```python
%%time
predicted_train = model1.predict(X_tf_idf_train)
```

```python
print('LogisticRegression. Метрика F1 на тренировочных данных: %.3f'%f1_score(y_train_upsampled, predicted_train))
```

```python
%%time
predicted_test = model1.predict(X_tf_idf_test)
```

```python
print('LogisticRegression. Метрика F1 на тестовых данных: %.3f'%f1_score(y_test, predicted_test))
```

Метрика на тестовых данных выше необходимого порога, без особого тюннинга гиперпараметров.

#### Оценим важность коэффициентов

```python
model1.coef_.shape
```

```python
df_feature_coef = pd.DataFrame({'coef':model1.coef_[0], 'feature_names':vectorizer.get_feature_names()}).sort_values(by = 'coef', ascending= False)

# Негативные слова. Class 1

TOP20_Class1 = df_feature_coef.head(20)
TOP20_Class1
```

```python
# Негативные слова. Class 0

TOP20_Class0 = df_feature_coef.tail(20).sort_values(by = 'coef', ascending =True )
TOP20_Class0
```

### Модель LGBMClassifier

#### Подготовим и обучим модель

```python
model2 = LGBMClassifier(n_jobs = cores, random_state = state)
```

```python
model2.get_params()
```

```python
%%time
model2.fit(X_tf_idf_train, y_train_upsampled)
```

#### Оценим метрику F1

```python
%%time
predicted_train = model2.predict(X_tf_idf_train)
```

```python
print('LGBMClassifier. Метрика F1 на тренировочных данных: %.3f'%f1_score(y_train_upsampled, predicted_train))
```

```python
%%time
predicted_test = model2.predict(X_tf_idf_test)
```

```python
print('LGBMClassifier. Метрика F1 на тестовых данных: %.3f'%f1_score(y_test, predicted_test))
```

#### Оценим важность коэффициентов

В случае с деревьями можем оценить только наиболее важные фичи для целевого признака, Class1, т.е негативные слова

```python
df_feature_coef = pd.DataFrame({'coef':model2.feature_importances_, 'feature_names':vectorizer.get_feature_names()}).sort_values(by = 'coef', ascending= False)
TOP20_Class1 = df_feature_coef.head(20)
TOP20_Class1
```
